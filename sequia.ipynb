```python
# Reemplace 'my_file.csv' con la ruta de tu archivo CSV
df = pd.read_csv('precipitacionsbarcelonadesde1786.csv')

# Ahora 'df' es un DataFrame, y puedes realizar operaciones sobre él.
df.head(40)  # Esto imprimirá las primeras 5 filas del DataFrame.
```

```python
# Asegurándonos de que 'Any' es el índice del DataFrame
df.set_index('Any', inplace=True)
```

```python
df.head()
```

```python
# Usando melt para remodelar el DataFrame
df_melt = df.reset_index().melt(id_vars='Any', var_name='Mes', value_name='Precipitacion')
```

```python
mapping = {'Precip_Acum_Gener': 1, 'Precip_Acum_Febrer': 2, 'Precip_Acum_Marc': 3, 'Precip_Acum_Abril': 4,
           'Precip_Acum_Maig': 5, 'Precip_Acum_Juny': 6, 'Precip_Acum_Juliol': 7, 'Precip_Acum_Agost': 8,
           'Precip_Acum_Setembre': 9, 'Precip_Acum_Octubre': 10, 'Precip_Acum_Novembre': 11, 'Precip_Acum_Desembre': 12}
df_melt['Mes'] = df_melt['Mes'].map(mapping)
```

```python
df_melt
```

```python
df_melt_sorted = df_melt.sort_values(by=['Any', 'Mes'])

# Opcional: si quieres restablecer el índice después de ordenar (esto eliminará la antigua columna de índices)
df_melt_sorted.reset_index(drop=True, inplace=True)

df_melt
```

```python
# Reemplace 'my_file.csv' con la ruta de tu archivo CSV
dft = pd.read_csv('temperaturesbarcelonadesde1780_format_long.csv')

# Ahora 'df' es un DataFrame, y puedes realizar operaciones sobre él.
dft.head(40)  # Esto imprimirá las primeras 5 filas del DataFrame.
```

```python
# Asegurándonos de que 'Any'/'año' y 'Mes'/'mes' son del mismo tipo en ambos DataFrames
df_melt['Any'] = df_melt['Any'].astype(dft['Any'].dtypes)
df_melt['Mes'] = df_melt['Mes'].astype(dft['Mes'].dtypes)
```

```python
# Usando merge para fusionar los DataFrames
df_combined = dft.merge(df_melt, how='left', left_on=['Any', 'Mes'], right_on=['Any', 'Mes'])
```

```python
df_combined
```

```python
# Asegúrate de que 'Any' es de tipo numérico para que la comparación funcione correctamente
df_combined['Any'] = df_combined['Any'].astype(int)

# Ahora eliminamos las filas para las cuales 'Any' está entre 1780 y 1785
df_combined = df_combined[(df_combined['Any'] < 1780) | (df_combined['Any'] > 1785)]
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Crear una columna de fecha combinando las columnas de año y mes
df_combined['Fecha'] = pd.to_datetime(df_combined['Any'].astype(str) + '-' + df_combined['Mes'].astype(str))


# Ordenar el DataFrame por la columna de fecha
data = df_combined.sort_values('Fecha')

# Crear la figura y los subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), sharex=True)

# Gráfico de temperatura
ax1.plot(data['Fecha'], data['Temperatura'], label='Temperatura')
ax1.set_ylabel('Temperatura')

# Gráfico de precipitación
ax2.plot(data['Fecha'], data['Precipitacion'], label='Precipitación')
ax2.set_ylabel('Precipitación')

# Títulos de los subplots
ax1.set_title('Temperatura por Mes')
ax2.set_title('Precipitación por Mes')

# Ajustar las etiquetas del eje x
fig.autofmt_xdate(rotation=45)

# Leyendas
ax1.legend()
ax2.legend()

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```
